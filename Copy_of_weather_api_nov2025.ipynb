```python
import requests

def get_weather(city):
   url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid=79007029562c2ab46617d801d8018750&units=metric"

   try:
       response= requests.get(url)
       weather_data = response.json()

       temperature = weather_data["main"]["temp"]
       humidity = weather_data["main"]["humidity"]
       max_temp = weather_data["main"]["temp_max"]
       min_temp = weather_data["main"]["temp_min"]
       wind_speed = weather_data["wind"]["speed"]
       #weather_description = weather_data["weather"]["description"]
       country = weather_data["sys"]["country"]


       print(f"Weather in {city}:")
       print(f"Temperature {temperature}¬∞C:")
       print(f"Humidity : {humidity}%")
       print(f"Max Temperature: {max_temp}¬∞C:")
       print(f"Min Temperature: {min_temp}¬∞C:")
       print(f"Wind Speed: {wind_speed} m/s:")
      #print(f"Weather description:{weather_description}")
       print(f"country:{country}:")

   except requests.exceptions.RequestException:
        print("Error occured")


#usage
city_name = input("Enter the name of city:")
get_weather(city_name)
```

```python
import gradio as gr
import requests

def get_weather(city):
    api_key = "79007029562c2ab46617d801d8018750"  # replace with your own key
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    response = requests.get(url)
    data = response.json()
    if data["cod"] == 200:
        temp = data["main"]["temp"]
        description = data["weather"][0]["description"]
        return f"The current temperature in {city} is {temp}¬∞C with {description}."
    else:
        return "Sorry, I couldn't find weather information for that city."

# Professional Gradio interface
iface = gr.Interface(
    fn=get_weather,
    inputs="text",
    outputs="text",
    title="üåç Global Weather App",
    description="Get real-time weather updates for any city worldwide.",
    theme="default",
    # Use a local file like 'logo.png' or a hosted image URL
    # Example: logo="https://upload.wikimedia.org/wikipedia/commons/e/e7/Weather_icon.png"
    # If you have a local logo file, place it in the same folder and use "logo.png"
    # Note: Gradio supports 'article' for extra info too
    examples=[["Abu Dhabi"], ["London"], ["New York"], ["Tokyo"],["Munich"]],
    allow_flagging="never"
)

iface.launch()
```
