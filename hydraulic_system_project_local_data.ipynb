## 2. Load Sensor Data from `data/`

The original dataset provides one **tab-delimited text file per sensor**.  
Typical filenames (you should have these in `data/`):

- `PS1.txt` … `PS6.txt` – pressure sensors  
- `EPS1.txt` – motor power  
- `FS1.txt`, `FS2.txt` – flow sensors  
- `TS1.txt` … `TS4.txt` – temperature sensors  
- `VS1.txt` – vibration  
- `CE.txt`, `CP.txt`, `SE.txt` – virtual / efficiency signals  

Each file:
- Rows = cycles (2205)  
- Columns = time points within the 60-second cycle  


```python
# Define expected sensor files
sensor_files = {
    "PS1": "PS1.txt",
    "PS2": "PS2.txt",
    "PS3": "PS3.txt",
    "PS4": "PS4.txt",
    "PS5": "PS5.txt",
    "PS6": "PS6.txt",
    "EPS1": "EPS1.txt",
    "FS1": "FS1.txt",
    "FS2": "FS2.txt",
    "TS1": "TS1.txt",
    "TS2": "TS2.txt",
    "TS3": "TS3.txt",
    "TS4": "TS4.txt",
    "VS1": "VS1.txt",
    "CE": "CE.txt",
    "CP": "CP.txt",
    "SE": "SE.txt",
}

def load_sensor_matrix(sensor_name, filename, data_dir=DATA_DIR):
    """Load a sensor .txt file (tab-delimited, no header) as DataFrame."""
    path = os.path.join(data_dir, filename)
    if not os.path.exists(path):
        raise FileNotFoundError(f"File not found: {path}. Please make sure it is in the data/ folder.")
    # Added on_bad_lines='skip' to handle rows with incorrect number of fields
    # Added low_memory=False to better handle mixed types and avoid DtypeWarning
    df = pd.read_csv(path, sep="\t", header=None, on_bad_lines='skip', low_memory=False)
    df.columns = [f"{sensor_name}_t{j}" for j in range(df.shape[1])]
    return df

# Load all sensors into a dict: name -> DataFrame
sensor_data = {}
for name, fname in sensor_files.items():
    print(f"Loading {name} from {fname}...")
    sensor_data[name] = load_sensor_matrix(name, fname)

# Check shapes (all should have same number of rows)
for name, df in sensor_data.items():
    print(f"{name}: shape = {df.shape}")
```

## 3. Load Target Condition Labels from `profile.txt`

The file `profile.txt` (tab-delimited) contains cycle-wise condition labels:

1. Cooler condition (%): 3, 20, 100  
2. Valve condition (%): 100, 90, 80, 73  
3. Internal pump leakage: 0, 1, 2  
4. Hydraulic accumulator pressure (bar): 130, 115, 100, 90  
5. Stability flag: 0 (stable), 1 (not stable)


```python
profile_path = os.path.join(DATA_DIR, "profile.txt")
if not os.path.exists(profile_path):
    raise FileNotFoundError(f"Targets file not found: {profile_path}")

y = pd.read_csv(
    profile_path,
    sep="\t",
    header=None,
    names=["cooler_condition", "valve_condition", "pump_leakage", "accumulator_pressure", "stable_flag"],
)

print("Targets shape:", y.shape)
display(y.head())

# Basic distribution of targets
for col in y.columns:
    print(f"\n=== {col} ===")
    print(y[col].value_counts())
```

## 4. From Raw Time-Series to Summary Features (`X_summary`)

For each sensor matrix (e.g., PS1), we compute **per-cycle summary features**:

- mean  
- standard deviation  
- min / max  
- median  
- 10th & 90th percentile  
- simple trend: `last_value − first_value`  

We then concatenate these summaries across all sensors to get a compact `X_summary`.


```python
def summarize_sensor_block(block_df, sensor_name):
    """Compute per-cycle summary stats for a sensor block (n_cycles × n_timepoints)."""
    # Ensure all data in the block_df is numeric, coercing errors to NaN
    block_df = block_df.apply(pd.to_numeric, errors='coerce')
    arr = block_df.to_numpy()
    # Use NaN-aware NumPy functions to handle potential NaNs introduced by coercion
    mean  = np.nanmean(arr, axis=1)
    std   = np.nanstd(arr, axis=1)
    minv  = np.nanmin(arr, axis=1)
    maxv  = np.nanmax(arr, axis=1)
    med   = np.nanmedian(arr, axis=1)
    p10   = np.nanpercentile(arr, 10, axis=1)
    p90   = np.nanpercentile(arr, 90, axis=1)
    trend = arr[:, -1] - arr[:, 0]

    summary = pd.DataFrame({
        f"{sensor_name}_mean": mean,
        f"{sensor_name}_std": std,
        f"{sensor_name}_min": minv,
        f"{sensor_name}_max": maxv,
        f"{sensor_name}_median": med,
        f"{sensor_name}_p10": p10,
        f"{sensor_name}_p90": p90,
        f"{sensor_name}_trend": trend,
    })
    return summary

# Build X_summary by concatenating summaries of all sensors
summary_parts = []
for sensor_name, df_sensor in sensor_data.items():
    summary = summarize_sensor_block(df_sensor, sensor_name)
    summary_parts.append(summary)

X_summary = pd.concat(summary_parts, axis=1)
print("X_summary shape:", X_summary.shape)
display(X_summary.head())
```

## 6. Baseline Model – Random Forest

We train a **Random Forest** classifier on `X_summary` to predict `pump_leakage`.


```python
# Align X_summary with the length of target (2205 rows)
X_summary_aligned = X_summary.loc[target.index]

X_train, X_test, y_train, y_test = train_test_split(
    X_summary_aligned, # Use the aligned X_summary
    target,
    test_size=0.2,
    random_state=42,
    stratify=target,
)

print("Aligned X_summary shape:", X_summary_aligned.shape)
print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
```

```python
rf_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("rf", RandomForestClassifier(
        n_estimators=300,
        max_depth=None,
        random_state=42,
        n_jobs=-1,
    )),
])

rf_pipeline.fit(X_train, y_train)
y_pred_rf = rf_pipeline.predict(X_test)

print(f"Random Forest – Classification report for target = {target_col}")
print(classification_report(y_test, y_pred_rf))
```

```python
# Confusion matrix for Random Forest
cm_rf = confusion_matrix(y_test, y_pred_rf, labels=np.sort(target.unique()))
disp_rf = ConfusionMatrixDisplay(confusion_matrix=cm_rf, display_labels=np.sort(target.unique()))
disp_rf.plot(values_format="d")
plt.title(f"Random Forest – Confusion Matrix ({target_col})")
plt.show()
```

## 7. Gradient Boosting with Learning Rate – XGBoost

Now we train **XGBoost**, which uses a `learning_rate` hyperparameter to control how fast it learns.


```python
xgb_pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("xgb", XGBClassifier(
        n_estimators=300,
        learning_rate=0.05,   # <-- learning rate
        max_depth=5,
        subsample=0.9,
        colsample_bytree=0.9,
        objective="multi:softmax" if len(np.unique(target)) > 2 else "binary:logistic",
        eval_metric="mlogloss" if len(np.unique(target)) > 2 else "logloss",
        random_state=42,
        n_jobs=-1,
        tree_method="hist",
    )),
])

xgb_pipeline.fit(X_train, y_train)
y_pred_xgb = xgb_pipeline.predict(X_test)

print(f"XGBoost – Classification report for target = {target_col}")
print(classification_report(y_test, y_pred_xgb))
```

```python
# Confusion matrix for XGBoost
cm_xgb = confusion_matrix(y_test, y_pred_xgb, labels=np.sort(target.unique()))
disp_xgb = ConfusionMatrixDisplay(confusion_matrix=cm_xgb, display_labels=np.sort(target.unique()))
disp_xgb.plot(values_format="d")
plt.title(f"XGBoost – Confusion Matrix ({target_col})")
plt.show()
```

```python
# Compare mean CV accuracy for RF vs XGBoost
rf_scores = cross_val_score(rf_pipeline, X_train, y_train, cv=5, n_jobs=-1)
xgb_scores = cross_val_score(xgb_pipeline, X_train, y_train, cv=5, n_jobs=-1)

print("Random Forest CV accuracy (mean):", rf_scores.mean())
print("XGBoost CV accuracy (mean):     ", xgb_scores.mean())
```

```python
import gradio as gr
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assume X_summary is already defined and available from previous cells

def perform_univariate_eda(feature_name):
    if feature_name not in X_summary.columns:
        return None, "Please select a valid feature."

    # Create a figure for the plot
    fig, ax = plt.subplots(figsize=(10, 6))

    # Plot histogram and KDE
    sns.histplot(X_summary[feature_name], kde=True, ax=ax)
    ax.set_title(f'Distribution of {feature_name}', fontsize=16)
    ax.set_xlabel(feature_name, fontsize=12)
    ax.set_ylabel('Count', fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.6)

    # Get descriptive statistics
    stats = X_summary[feature_name].describe().to_string()

    # Important: close the plot to prevent it from displaying twice in some environments
    plt.close(fig)

    return fig, stats

# Get all feature names from X_summary for the dropdown
feature_choices = X_summary.columns.tolist()

# Create Gradio Interface
eda_iface = gr.Interface(
    fn=perform_univariate_eda,
    inputs=[
        gr.Dropdown(feature_choices, label="Select Feature for Univariate EDA", value=feature_choices[0] if feature_choices else None)
    ],
    outputs=[
        gr.Plot(label="Distribution Plot"),
        gr.Textbox(label="Descriptive Statistics")
    ],
    title="Univariate Exploratory Data Analysis (EDA)",
    description="Select a feature from the summary dataset to view its distribution and descriptive statistics."
)

# Launch the interface
eda_iface.launch(debug=True, share=True)
```

```python
import gradio as gr
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assume X_summary is already defined and available from previous cells

def perform_bivariate_eda(feature1_name, feature2_name):
    if feature1_name not in X_summary.columns or feature2_name not in X_summary.columns:
        return None, "Please select two valid features.", ""

    # Create a figure for the plot
    fig, ax = plt.subplots(figsize=(10, 7))

    # Plot scatter plot
    sns.scatterplot(x=X_summary[feature1_name], y=X_summary[feature2_name], ax=ax, alpha=0.6)
    ax.set_title(f'Bivariate Distribution: {feature1_name} vs {feature2_name}', fontsize=16)
    ax.set_xlabel(feature1_name, fontsize=12)
    ax.set_ylabel(feature2_name, fontsize=12)
    ax.grid(True, linestyle='--', alpha=0.6)

    # Calculate Pearson correlation
    correlation = X_summary[feature1_name].corr(X_summary[feature2_name])
    correlation_text = f"Pearson Correlation: {correlation:.4f}"

    # Important: close the plot to prevent it from displaying twice in some environments
    plt.close(fig)

    return fig, correlation_text, ""

# Get all feature names from X_summary for the dropdowns
feature_choices = X_summary.columns.tolist()

# Create Gradio Interface
bivariant_eda_iface = gr.Interface(
    fn=perform_bivariate_eda,
    inputs=[
        gr.Dropdown(feature_choices, label="Select Feature 1", value=feature_choices[0] if feature_choices else None),
        gr.Dropdown(feature_choices, label="Select Feature 2", value=feature_choices[1] if len(feature_choices) > 1 else None)
    ],
    outputs=[
        gr.Plot(label="Scatter Plot"),
        gr.Textbox(label="Correlation"),
        gr.Textbox(label="Notes", interactive=False)
    ],
    title="Bivariate Exploratory Data Analysis (EDA)",
    description="Select two features from the summary dataset to visualize their relationship and calculate correlation."
)

# Launch the interface
bivariant_eda_iface.launch(debug=True, share=True)
```

```python
def simulate_leakage(input_features: dict):
    """Simulates pump leakage prediction given a dictionary of feature values.

    Args:
        input_features (dict): A dictionary where keys are feature names
                               and values are their corresponding input values.

    Returns:
        tuple: A tuple containing the predicted class (0, 1, or 2)
               and a dictionary of class probabilities.
    """
    # Create a DataFrame for a single sample, initialized with training data means
    # X_train_means should be available from previous Gradio cell if executed, or calculated here
    if 'X_train_means' not in globals():
        X_train_means = X_train.mean().to_dict()

    sample_df = pd.DataFrame([X_train_means])

    # Update specified features with input_features
    for feature, value in input_features.items():
        if feature in sample_df.columns:
            sample_df[feature] = value
        else:
            print(f"Warning: Input feature '{feature}' not found in model features. Skipping.")

    # Make prediction using the trained XGBoost pipeline
    prediction = xgb_pipeline.predict(sample_df)[0]
    probabilities = xgb_pipeline.predict_proba(sample_df)[0]

    # Map numerical prediction to a more readable format
    leakage_map = {0: "No Leakage (0)", 1: "Medium Leakage (1)", 2: "High Leakage (2)"}
    predicted_class_name = leakage_map.get(prediction, f"Unknown ({prediction})")

    prob_dict = {
        "No Leakage (0)": probabilities[0],
        "Medium Leakage (1)": probabilities[1],
        "High Leakage (2)": probabilities[2]
    }

    return predicted_class_name, prob_dict

print("Simulation function 'simulate_leakage' is ready.")

# Example usage (you can uncomment and run this to test):
# example_input = {
#     'SE_p90': 70.0,  # A value for the most important feature
#     'SE_median': 65.0, # A value for the second most important feature
#     'FS1_mean': 30.0 # Another example feature
# }
# predicted_class, probs = simulate_leakage(example_input)
# print(f"\nPredicted Class: {predicted_class}")
# print(f"Probabilities: {probs}")
```

```python
import gradio as gr
import pandas as pd
import numpy as np

# Ensure X_train_means is available (defined in previous Gradio cell or compute if not)
if 'X_train_means' not in globals():
    X_train_means = X_train.mean().to_dict()

def simulate_leakage_with_specific_features(
    se_median: float,
    se_p90: float,
    fs1_median: float,
    eps1_mean: float,
    ts2_min: float
):
    """Simulates pump leakage prediction given specific feature values.

    Args:
        se_median (float): Value for 'SE_median'
        se_p90 (float): Value for 'SE_p90'
        fs1_median (float): Value for 'FS1_median'
        eps1_mean (float): Value for 'EPS1_mean'
        ts2_min (float): Value for 'TS2_min'

    Returns:
        tuple: A tuple containing the predicted class (0, 1, or 2)
               and a dictionary of class probabilities.
    """
    # Create a DataFrame for a single sample, initialized with training data means
    sample_df = pd.DataFrame([X_train_means])

    # Update the specified features with the input values
    sample_df['SE_median'] = se_median
    sample_df['SE_p90'] = se_p90
    sample_df['FS1_median'] = fs1_median
    sample_df['EPS1_mean'] = eps1_mean
    sample_df['TS2_min'] = ts2_min

    # Make prediction using the trained XGBoost pipeline
    prediction = xgb_pipeline.predict(sample_df)[0]
    probabilities = xgb_pipeline.predict_proba(sample_df)[0]

    # Map numerical prediction to a more readable format
    leakage_map = {0: "No Leakage (0)", 1: "Medium Leakage (1)", 2: "High Leakage (2)"}
    predicted_class_name = leakage_map.get(prediction, f"Unknown ({prediction})")

    prob_dict = {
        "No Leakage (0)": probabilities[0],
        "Medium Leakage (1)": probabilities[1],
        "High Leakage (2)": probabilities[2]
    }

    return predicted_class_name, prob_dict


# Create Gradio Interface for the specific simulation function
simulation_iface = gr.Interface(
    fn=simulate_leakage_with_specific_features, # Use the function defined here
    inputs=[
        gr.Slider(minimum=X_train['SE_median'].min(), maximum=X_train['SE_median'].max(), value=X_train_means['SE_median'], label="SE_median"),
        gr.Slider(minimum=X_train['SE_p90'].min(), maximum=X_train['SE_p90'].max(), value=X_train_means['SE_p90'], label="SE_p90"),
        gr.Slider(minimum=X_train['FS1_median'].min(), maximum=X_train['FS1_median'].max(), value=X_train_means['FS1_median'], label="FS1_median"),
        gr.Slider(minimum=X_train['EPS1_mean'].min(), maximum=X_train['EPS1_mean'].max(), value=X_train_means['EPS1_mean'], label="EPS1_mean"),
        gr.Slider(minimum=X_train['TS2_min'].min(), maximum=X_train['TS2_min'].max(), value=X_train_means['TS2_min'], label="TS2_min")
    ],
    outputs=[
        gr.Textbox(label="Predicted Pump Leakage Class"),
        gr.Label(label="Class Probabilities")
    ],
    title="Hydraulic System Pump Leakage Simulation (Top 5 Features)",
    description="Adjust the values of the top 5 influential features to simulate pump leakage prediction."
)

# Launch the interface
simulation_iface.launch(debug=True, share=True)
```
